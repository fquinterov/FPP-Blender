```python
import matplotlib.pyplot as plt;
import numpy as np;
import math;
import cv2

resx = 1280 # Projector x resolution
resy = 800  # Projector y resolution

pitch = [1280, 640, 320, 160, 80, 40, 20] # Pringe pitch

phase = [0.0, -2*math.pi/3, 2*math.pi/3] # The phase must be a value in radians (e.g. 2 * math.pi / 3)
```

### 1. Vertical stripes

```python
i = np.arange(resx) # Sequential vector with coordinates in pixels
i = i - resx/2

# Here is the equation for generation of fringe patterns
for p in range(len(pitch)):
    nphase = 0
    for ph in range(len(phase)):
        nphase = nphase + 1
        I = np.uint8((255/2)*(1+np.cos(2.0*math.pi*i/pitch[p] + phase[ph])))
        X = np.tile(I,(resy,1))
        cv2.imwrite("./v_"+str(pitch[p])+"_"+str(nphase)+".bmp", X)

        plt.imshow(X, cmap = "gray")
        plt.title("Pitch = " + str(pitch[p]) + ", Phase = " + str(phase[ph]))
        plt.show()
```

### 2. Horizontal stripes

```python
j = np.arange(resy) # Sequential vector with coordinates in pixels
j = j - resy/2

# Here is the equation for generation of fringe patterns
for p in range(len(pitch)):
    nphase = 0
    for ph in range(len(phase)):
        nphase = nphase + 1
        I = np.uint8((255/2)*(1+np.cos(2.0*math.pi*j/pitch[p] + phase[ph])))
        I = np.matrix(I)
        X = np.tile(I.T,(1,resx));
        cv2.imwrite("./h_"+str(pitch[p])+"_"+str(nphase)+".bmp", X)
```
